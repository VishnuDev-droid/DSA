##### Brute force solution: 
Steps:
1. start with pos variable, check list empty, if empty return -1
2. pos ==0, check arr[pos], if found, return pos, else increment pos by 1
3. repeat step 2


```python
import time
def locate_query_elem_brute_force(arr, query):
    """Returns 1st occerance of position of query element 
    arr: List
    query: float"""
    
    pos = 0
    if not len(arr):
        return -1
    for pos in range(len(arr)):
        if arr[pos] == query:
            return pos
    return -1
```

#### Optimized Soln1 : Binary search
Steps:
1. Find middle element
2. Compare middle elem with query elem
3. If query elem < middle elem, remove right portion of list
4. If query elem > middle elem, remove left portion of list
5. Repeat 3, 4 until we get query position

```python
test = {
    'input': {
        'arr': [40, 30, 20, 18, 16, 13, 11, 10, 7, 4, 3, 1, 0],
        'query': 1
    },
    'output': 6
}

def locate_query_elem_binary_search(arr, query):
    """Returns 1st occerance of position of query element 
    arr: List
    query: float"""
    low = 0
    high = len(arr) - 1
    
    while low <= high:
        print('**')
        middle_pos = (low + high)//2
        middle_elem = arr[middle_pos]
        if query > middle_elem:
            high = middle_pos-1
        elif query < middle_elem:
            low = middle_pos+1
        elif middle_elem == query:
            if middle_pos-1 >= 0 and arr[middle_pos-1] == query:
                high = middle_pos-1
            else:
                return middle_pos
    return -1
            
        
```

#### create a data structure which can store 100 million records and perform insertion, search, update and list operations efficiently.

```python
class User:
    def __init__(self, username, name, user_email):
        self.username = username
        self.name = name
        self.email = user_email
        print(f"User {self.username} Created")
    def __repr__(self):
        return f'User(Username: {self.username}, Name: {self.name}, Email: {self.email})'
    def __str__(self):
        return self.__repr__()
    
class UserDataBase:
    def __init__(self):
        """constructor for UserDataBase"""
        self.users = []
    
    def insert(self, user):
        """Add user to database"""
        i = 0
        while i < len(self.users):
            if self.users[i].username > user.username:
                break
            i+=1
        self.users.insert(i, user)
            
    def find(self, username):
        """Find a user from database using username"""
        for user in self.users:
            if user.username == username:
                return user
    
    def update(self, user):
        """Update user profile"""
        target_user = self.find(user.username)
        target_user.name, target_user.email = user.name, user.email

    def list_all(self):
        """List all User profiles"""
        return self.users
```

##### Binary search Tree
Properties:
1. lexicographically smaller on left side nodes, lexocographically higher on right nodes
2. treemap (Associated with keys & values)
3. Balanced (Does not skew on one side heavly)
4. Height of a tree (Levels)
5. For balanced tree:
        Level0 : 1 node
        Level1: 2 nodes
        Level2: 4 nodes
        Level3: 8 nodes, etc..
        Level k-1: 2^k-1 nodes
6. Relation bw num nodes and level
        N = 1 + 2 ^1 + 2 ^2 + 2 ^ 3 + .... + 2 ^k-1
        
        N + 1 = 2 ^ k
        k (Height of a tree) = log(N+1) <= log(N) + 1

#### Traverse In-order

```python
def traverse_inorder(node):
    if node is None:
        return []
    return traverse_inorder(node.left) + [node.key] + traverse_inorder(node.right)
traverse_inorder(binary_tree)
```

```python
def traverse_pre_order(node):
    if node is None:
        return []
    return [node.key] + traverse_pre_order(node.left) + traverse_pre_order(node.right)
traverse_pre_order(binary_tree)
```

```python
def tree_height(node):
    if node is None:
        return 0
    return 1 + max(tree_height(node.left), tree_height(node.right))
tree_height(binary_tree)
```

```python
def tree_size(node):
    if node is None:
        return 0
    return 1 + tree_size(node.left) + tree_size(node.right)
tree_size(binary_tree)
```

```python
class TreeNode():
    def __init__(self, key):
        self.key, self.left, self.right = key, None, None
    
    def height(self):
        if self is None:
            return 0
        return 1 + max(TreeNode.height(self.left), TreeNode.height(self.right))
    
    def size(self):
        if self is None:
            return 0
        return 1 + TreeNode.size(self.left) + TreeNode.size(self.right)

    def traverse_in_order(self):
        if self is None: 
            return []
        return (TreeNode.traverse_in_order(self.left) + 
                [self.key] + 
                TreeNode.traverse_in_order(self.right))
    
    def display_keys(self, space='\t', level=0):
        # If the node is empty
        if self is None:
            print(space*level + '∅')
            return   

        # If the node is a leaf 
        if self.left is None and self.right is None:
            print(space*level + str(self.key))
            return

        # If the node has children
        self.display_keys(self.right, space, level+1)
        print(space*level + str(self.key))
        self.display_keys(self.left,space, level+1)    
    
    def to_tuple(self):
        if self is None:
            return None
        if self.left is None and self.right is None:
            return self.key
        return TreeNode.to_tuple(self.left),  self.key, TreeNode.to_tuple(self.right)
    
    def __str__(self):
        return "BinaryTree <{}>".format(self.to_tuple())
    
    def __repr__(self):
        return "BinaryTree <{}>".format(self.to_tuple())
    
    @staticmethod    
    def parse_tuple(data):
        if data is None:
            node = None
        elif isinstance(data, tuple) and len(data) == 3:
            node = TreeNode(data[1])
            node.left = TreeNode.parse_tuple(data[0])
            node.right = TreeNode.parse_tuple(data[2])
        else:
            node = TreeNode(data)
        return node
    @staticmethod
    def remove_none(nums):
        return [x for x in nums if x is not None]
    @staticmethod
    def is_BST(node):
        """Function returns is bst flag, min key, max key """
        if node is None:
            return True, None, None
        is_bst_l, min_l, max_l = TreeNode.is_BST(node.left)
        is_bst_r, min_r, max_r = TreeNode.is_BST(node.right)
        
        print('is_bst_l: ', is_bst_l, node.key)
        print('is_bst_r: ', is_bst_r, node.key)
        is_bst_node = (is_bst_l and is_bst_r and 
                            (max_l is None or max_l < node.key ) and
                            (min_l is None or node.key < min_l))
        min_key = min(TreeNode.remove_none([min_l, node.key, min_r]))
        max_key = max(TreeNode.remove_none([max_l, node.key, max_r]))

        return is_bst_node, min_key, max_key
```

#### Check a binary tree a binary search tree (BST)
Property to be satisfied:
1. The left subtree of any node only contains nodes with keys less than the node's key
2. The right subtree of any node only contains nodes with keys greater than the node's key


```python
def remove_none(nums):
    return [x for x in nums if x is not None]

def is_BST(node):
    """Function returns is bst flag, min key, max key """
    if node is None:
        return True, None, None
    is_bst_l, min_l, max_l = is_BST(node.left)
    is_bst_r, min_r, max_r = is_BST(node.right)

    is_bst_node = (is_bst_l and is_bst_r and 
                        (max_l is None or max_l < node.key ) and
                        (min_l is None or node.key < min_l))
    min_key = min(remove_none([min_l, node.key, min_r]))
    max_key = max(remove_none([max_l, node.key, max_r]))
    
    return is_bst_node, min_key, max_key
# is_BST(binary_tree)
```
